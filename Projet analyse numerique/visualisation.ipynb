```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import time

def simulate_epidemic():

    # Paramtres de la simulation
    num_individuals = 100  # Nombre total d'individus
    num_infected = 1       # Nombre initial d'individus infects
    simulation_time = 100  # Dure de la simulation

    # Paramtres du systme diffrentiel
    beta = 0.3   # Taux de transmission
    alpha = 0.1  # Taux de rcupration
    gamma = 0.05 # Taux de transition de I  T
    eta = 0.02   # Taux de transition de T  R
    delta = 0.1  # Facteur pour le modle de dlai

    # Conditions initiales du systme diffrentiel
    S = 0.99
    I = 0.01
    T = 0
    R = 0

    # Crer la figure et dfinir les axes
    fig, ax = plt.subplots()
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_xlabel('Position X')
    ax.set_ylabel('Position Y')
    ax.set_title('Simulation d\'pidmie')

    # Initialiser les positions et les tats des individus
    positions = 10 * np.random.rand(num_individuals, 2)  # Positions alatoires
    states = np.zeros(num_individuals, dtype=int)         # 0: Sain, 1: Infect, 2: Expos, 3: Rcupr, 4: Dcd
    states[:num_infected] = 1                            # Dfinir les premiers individus comme infects

    # Couleurs pour les diffrents tats
    colors = ['b', 'r', 'm', 'g', 'k']  # Bleu: Sain, Rouge: Infect, Magenta: Expos, Vert: Rcupr, Noir: Dcd

    # Dessiner les cercles pour chaque individu
    circles = []
    for i in range(num_individuals):
        circle = patches.Circle(positions[i], radius=0.25, color=colors[states[i]])
        ax.add_patch(circle)
        circles.append(circle)

    # Boucle de simulation
    for t in range(simulation_time):
        # Mettre  jour les tats des individus en fonction des valeurs du systme diffrentiel
        dS = -beta * S * I
        dI = beta * S * I - (alpha + gamma) * I
        dT = gamma * I - eta * T
        dR = eta * T
        
        # Mise  jour des populations
        S += dS
        I += dI
        T += dT
        R += dR
        
        # Mettre  jour les tats des individus en fonction des nouvelles valeurs du systme diffrentiel
        for i in range(num_individuals):
            # Dplacer les individus alatoirement
            positions[i] += np.random.randn(2) * 0.1

            # Vrifier les collisions avec les bords de la zone
            positions[i, 0] = np.clip(positions[i, 0], 0, 10)
            positions[i, 1] = np.clip(positions[i, 1], 0, 10)
            
            # Dfinir l'tat en fonction des valeurs des populations
            if np.random.rand() < S:
                states[i] = 0  # Sain
            elif np.random.rand() < S + I:
                states[i] = 1  # Infect
            elif np.random.rand() < S + I + T:
                states[i] = 2  # Expos
            elif np.random.rand() < S + I + T + R:
                states[i] = 3  # Rcupr
            else:
                states[i] = 4  # Dcd

            # Mettre  jour la couleur et la position du cercle en fonction de l'tat
            circles[i].set_color(colors[states[i]])
            circles[i].center = positions[i]

        # Mettre  jour l'affichage
        plt.draw()
        plt.pause(0.1)  # Pause pour ralentir la simulation

    plt.show()

simulate_epidemic()
```
